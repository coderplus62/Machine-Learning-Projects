```python
import numpy as np
import pandas as pd 
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
file = "creditcard.csv"
data = pd.read_csv(file)
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data["Class"].value_counts()
```

```python
legit = data[data.Class == 0]
fraud = data[data.Class == 1]

print(legit.shape)
print(fraud.shape)
```

```python
legit = legit.sample(n=492)
print(legit.shape)
print(fraud.shape)
```

```python
update_data = pd.concat([legit, fraud], axis = 0)
update_data.head()
```

```python
update_data["Class"].value_counts()
```

```python
X = update_data.drop(columns="Class", axis = 1)
y = update_data["Class"]

print(X)
print(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

print("Features")
print(X.shape)
print(X_train.shape)
print(X_test.shape)

print("\nTarget")
print(y.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
train_prediction = classifier.predict(X_train)
train_accuracy = accuracy_score(train_prediction, y_train)*100

test_prediction = classifier.predict(X_test)
test_accuracy = accuracy_score(test_prediction, y_test)*100

print("\n\n========== Model from Sklearn API ==========\n\n")

print(f"Accuracy on train data: {train_accuracy}%")
print(f"Accuracy on test data: {test_accuracy}%")
```
